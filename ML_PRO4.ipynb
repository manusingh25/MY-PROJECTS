# DATA PREAPRATION

- we have to normalise the variables to express them in the same range of values. In other words, normalizing means adjusting values measured on different scales to a common scale 

```python
# data normalization

from sklearn.preprocessing import normalize 
scaled =normalize(df)
scaled=pd.DataFrame(scaled)
scaled.head()
```

```python
from sklearn.cluster import KMeans
import warnings 
warnings.filterwarnings('ignore')
```

```python
sse=[]
for i in range(1,5):
    Kmeans=KMeans(n_clusters=i,init='random',random_state=42)
    Kmeans.fit(df)
    sse.append(Kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,5),sse)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('clustering score')
plt.show()
```

```python
sse
```

```python
# fit the method and predict 

Kmeans =KMeans(n_clusters=2,random_state=42)
Kmeans.fit(df)
```

```python
pred=Kmeans.predict(df)
pred
```

```python
Kmeans.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled,pred)
```

# HIERARICHAL CLUSTERING

```python
import pandas as pd 
```

```python
df=pd.read_csv("C:/Users/Manu/Downloads/winequality_red.csv")
df
```

# DATA PREPARATION 

```python
from sklearn.preprocessing import normalize
scaled=normalize(df)
scaled=pd.DataFrame(scaled)
scaled.head()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,7))
dendrogram=sch.dendrogram(sch.linkage(scaled,method='ward'))

plt.title('Dendrogram')
plt.xlabel('Quality of wine')
plt.ylabel('apply all featuers')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
pred=cluster.fit_predict(scaled)
pred
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled,pred)
```

# apply minmax sclaer with the help of KMeans and Hierarichal clustering

```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv("C:/Users/Manu/Downloads/winequality_red.csv")
df
```

```python
df.head()
```

```python
z=df.columns
z
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms
```

```python
df=ms.fit_transform(df)
df
```

```python
df=pd.DataFrame(df)
df
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2,random_state=0)
km.fit(df)
```

```python
sse=[]
k_rng=range(1,5)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of sqaured error')
plt.plot(k_rng,sse)
plt.show()
```

```python
km.cluster_centers_
```

```python
pred=km.predict(df)
pred
```

```python
km.inertia_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df,pred)
```
