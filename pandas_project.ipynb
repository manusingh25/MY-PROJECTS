```python
# making a pie chart 

plt.figure(figsize=(7,7))
plt.pie(list(a['winner'].value_counts()[0:5]),labels=list(a['winner'].value_counts()[0:5].keys()),autopct='%0.1f%%',colors=['blue','red','pink','yellow','teal'])
plt.show()
```

```python
c=a['winner'].value_counts()[0:5].keys()
d=a['winner'].value_counts()[0:5].value_counts()
```
