```python
from sklearn.cluster import KMeans
Kmeans = KMeans(n_clusters=2, random_state=0)

Kmeans.fit(z)
```

```python
sse=[]
k_rng=range(1,10)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(z[['status_type','num_reactions','num_comments','num_shares','num_likes','num_loves']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared error')
plt.plot(k_rng,sse)
```
